```python
df = pd.read_csv ('data.csv')
df.head()
```

```python
df = df[df["country"].str.contains("(total)")==False]
```

```python
df["country"] = df["country"].replace("United States Of America", "USA", regex=True)
```

```python
fig = px.choropleth(df, locations="country", color="rate_14_day", 
                    locationmode='country names',
                    animation_frame='year_week',
                    range_color=[0,100],
                    color_continuous_scale=px.colors.sequential.OrRd
                   )

fig.update_layout(title='Data on 14-day notification rate of new COVID-19 cases and deaths. Source: ECDC')
fig.show()
```
